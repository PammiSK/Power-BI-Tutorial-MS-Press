````python
# data = data given in via text input

import os
import pandas as pd

def store_sales_data(store_data):
    clean_data = [cd for cd in store_data.split("\n") if cd]
    
    timestamp = " ".join(clean_data[-1].split()[:-1])
    timestamp = timestamp.replace("/", "_")
    timestamp = timestamp.replace(":", "_")
    store = " ".join(clean_data[-2].split()[2:])
    
    sales_data = [sale.split() for sale in clean_data if sale.endswith("%")]
    total = sales_data.pop()

    for sd in sales_data:
        if len(sd) > 8:
            sales_data[sales_data.index(sd)] = [" ".join(sd[:-7])] + sd[-7:]
    
    column_headers = ["Sales Person", "Retail $ Incl Disc", "Discount", "Returns", "Net $", "Avg Sale Units", "Avg Sale $", "% Store Total"]
    df = pd.DataFrame(sales_data, columns = column_headers)
    
    columns_to_clean = ["Retail $ Incl Disc", "Discount", "Returns", "Net $", "Avg Sale Units", "Avg Sale $"]
    df[columns_to_clean] = df[columns_to_clean].replace(",", "", regex=True).astype(float)
    df["% Store Total"] = df["% Store Total"].str.rstrip("%").astype(float)
    
    df.to_csv(f"{store} {timestamp}.csv", index=False)
    return df
````
